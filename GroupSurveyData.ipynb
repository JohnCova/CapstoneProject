***#2: Find the low, medium, and maximum numbers of Earth by Country.***

- The country has minimum numbers of earths is Lebanon: 0.8
- The country has maximum numbers of Earths is United Kingdom: 7.3

```python
survey_df.groupby("Country_Currently_Residing_In")["Number_of_earths"].min().sort_values()
```

```python
survey_df.groupby("Country_Currently_Residing_In")["Number_of_earths"].min().plot.barh()
```

```python
survey_df.groupby("Country_Currently_Residing_In")["Number_of_earths"].mean().round(2).sort_values()
```

```python
survey_df.groupby("Country_Currently_Residing_In")["Number_of_earths"].mean().plot.barh()
```

```python
datamax = survey_df.groupby("Country_Currently_Residing_In")["Number_of_earths"].max()
datamax.sort_values()
```

```python
survey_df.groupby("Country_Currently_Residing_In")["Number_of_earths"].max().plot.barh()
```
